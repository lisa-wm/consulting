```python
import pandas as pd
import numpy as np
from ipynb.fs.full.helper_functions import normalize_tweet, build_feature_matrix, prepare_arrays, get_dfm
```

```python
# Load and prepare data

tweets_df = pd.read_excel("D:/UNI/01_Master/3. Semester/Consulting/consulting/2_code/labeling_lisa.xlsx")

# Keep labeled observations only
df = tweets_df[tweets_df['label'].notna()] 
tweets_df['label_binary'] = [1 if sentiment == 'positive' else 0 for sentiment in tweets_df.label]
tweets_df.head()
tweets_df.shape
```

```python
# Make it ready for training

tweets_train, labels_train = prepare_arrays(tweets_df)
```

```python
# Normalize the tweets

normalized_tweets = normalize_tweet(tweets_train, lemmatize=True, 
                     only_text_chars=True,
                     tokenize=False, 
                     stopwords_removal = True,
                     lower_case = False,
                     umlauts_removal = True)
# normalized_tweets[1]
len(normalized_tweets)
```

```python
# Do feature extraction (i.e. TF-IDF, bigram)

vectorizer, features = build_feature_matrix(documents=normalized_tweets,
                                                  feature_type='tfidf',
                                                  ngram_range=(2, 2), 
                                                  min_df=0.0, max_df=1.0) 
```

```python
# How does the resulting dtm look like? 

print(features.shape)
get_dfm(features = features.todense(), names = vectorizer.get_feature_names())
```
